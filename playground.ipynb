```python
from datasets import load_dataset

ds = load_dataset("AfterQuery/App-Bench")
```

```python
pandas_dfs = {}
for split_name, dataset in ds.items():
    pandas_dfs[split_name] = dataset.to_pandas()
```

```python
df = pandas_dfs['train']
```
