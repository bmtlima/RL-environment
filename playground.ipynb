```python
# Adding a simple example to data/prompts.csv:

new_data = {
    "#": [0],
    "App Name": ["Click Counter"],
    "App Description": ["A minimal web page with a button that increments a displayed counter."],
    "Prompt": [
        'Create a simple web application with a centered button labeled "Click Me". Above the button, display a counter that starts at 0. Every time the user clicks the button, the counter should increase by 1. Use React `useState` for state management.',
    ],
    "Addition for CLI Tools": [
        "You are working in a Next.js 16 environment with Tailwind CSS. Implement the solution in `app/page.tsx`. Do NOT create an index.html file. Ensure the UI is centered using Tailwind classes."
    ],
    "Rubric": [
        "1. Page displays a visible counter starting at 0.\n2. Page displays a clickable button.\n3. Clicking the button increments the counter value immediately."    ],
}

# 2. Create a DataFrame for the new row
new_df = pd.DataFrame(new_data)

# 3. Read the existing CSV

df = pd.read_csv("data/prompts.csv")

# 4. Concatenate the new row at the top (new_df first, then df)
combined_df = pd.concat([new_df, df], ignore_index=True)

# 5. Save the updated DataFrame back to CSV
combined_df.to_csv("data/prompts.csv", index=False)

print("Success! New row added to the top.")
print(combined_df.head(2))
```
